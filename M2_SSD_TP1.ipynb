# Imports et lecture donn√©es

```python
import pandas as pd
import numpy as np

data = pd.read_csv("Datagenus.csv", sep="\t", decimal=',')
```

```python
data
```

Calcul de Y (treedensity)

```python
treedensity = (
    data
    [[f"gen{i}" for i in range(1, 27 +1)]]
    .sum(axis="columns")
    /data["surface"]
)
```

```python
treedensity
```

Calcul des carr√©s des variables g√©ographiques quantitatives

```python
geography_quant = (
    data
     [
         ["lon", "lat", "altitude", "pluvio_yr"]
         + [f'pluvio_{i}' for i in range(1, 12 +1)]
     ]
)
```

```python
geography_quant
```

```python
geography_squares = (
  (geography_quant**2)
  .rename({
    col_name : f"squared_{col_name}"
    for col_name in geography_quant.columns
  }, axis="columns")
)
```

```python
geography_squares
```

Indicatrices de la variable

```python
geology_categories = data["geology"].unique()
geology_ind = pd.DataFrame({
    f"geology_cat{cat}": (data["geology"] == cat).astype(float)
    for cat_id, cat in enumerate(geology_categories)
})
```

```python
geology_categories
```

```python
geology_ind
```

```python
features = (
    pd.concat(
        [
            geography_quant,
            geography_squares,
            geology_ind,
            data[[f"evi_{i}" for i in range(1, 23 +1)]]
        ],
        axis="columns"
    )
)
```

2. Premi√®re r√©gression sur composantes principales:

```python
import sklearn
```

```python
pca_features = sklearn.decomposition.PCA().fit_transform(features)
pca_features
```

4. R√©gression PLS:

```python
X = features.to_numpy()
Y = treedensity.to_numpy()
```

```python
print(f"Dimensions de X : {X.shape}")
print(f"Dimensions de Y : {Y.shape}")
```

```python
n_train = 900
n_folds = 9
fold_size = 100
```

```python
X_train = X[:n_train]
Y_train = Y[:n_train]
```

```python
result = {n_components: [] for n_components in range(2, 30 +1)}

for fold in range(n_folds):
  fold_start_index = fold*fold_size
  fold_end_index = (fold+1)*fold_size

  for n_components in range(2, 30 +1):
    pls_regression = sklearn.cross_decomposition.PLSRegression(n_components = n_components)

    X_fold_out = np.concatenate((X_train[:fold_start_index], X_train[fold_end_index:]))
    X_fold = X_train[fold_start_index:fold_end_index]

    Y_fold_out = np.concatenate((Y_train[:fold_start_index], Y_train[fold_end_index:]))
    Y_fold = Y_train[fold_start_index:fold_end_index]

    pls_regression.fit(X_fold_out, Y_fold_out)
    Y_hat_fold = pls_regression.predict(X_fold)

    metric = sklearn.metrics.mean_squared_error(Y_fold, Y_hat_fold)
    result[n_components].append(metric)

result_per_n_components = {
    n_components: sum(result[n_components])/n_folds
    for n_components in range(2, 30 +1)
}
result_per_n_components
```

```python
n_components_optimal = list(result_per_n_components.keys())[np.argmin(list(result_per_n_components.values()))]
```

```python
n_components_optimal
```

```python
X_test = X[n_train:]
Y_test = Y[n_train:]
```

```python
pls_regression = sklearn.cross_decomposition.PLSRegression(n_components = n_components_optimal)
pls_regression.fit(X_train, Y_train)
Y_hat_test = pls_regression.predict(X_test)
sklearn.metrics.mean_squared_error(Y_test, Y_hat_test)
```
